```python
# Import requsit libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load your data
nir_data = pd.read_csv('NIRExtAll.csv')
uv_data = pd.read_csv('UVExtAll.csv')
vis_data = pd.read_csv('VISExtAll.csv')
```

```python
# Function to perform PCA and plot
def perform_pca_and_plot(data, title):
    # Descriptive statistics
    desc_stats = data.groupby(['Solvent', 'Shodhan']).describe()

    # PCA
    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(data.iloc[:, 4:])
    pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
    pca_df = pd.concat([pca_df, data[['Solvent', 'Shodhan']]], axis=1)

    # Plotting
    sns.scatterplot(x='PC1', y='PC2', hue='Solvent', style='Shodhan', data=pca_df)
    plt.title(title)
    plt.show()

    # Correlation matrix
    correlation_matrix = data.iloc[:, 4:].corr()
    print(correlation_matrix)
```

```python
# Perform PCA and plot for each dataset
perform_pca_and_plot(nir_data, 'NIR Segment PCA')
```

```python
perform_pca_and_plot(uv_data, 'UV Segment PCA')
```

```python
perform_pca_and_plot(vis_data, 'VIS Segment PCA')
```
